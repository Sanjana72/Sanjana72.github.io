```python
data=pd.read_csv('Ass5.csv')
data
```

```python
# Step:1) Clean the data
#     1.Remove the patient P006
# Getting the the index of patient P006
index=data[(data.PatientID=='P006')].index

# Dropping the row using index
data=data.drop(index)
data
```

```python
# Replace the NAN for the following two rows

# Using the approach in the ppt

data.loc[10:10,'Age'].fillna(62, inplace=True)

# Another way to directly update the record

data.loc[11,'Age']=89
data.loc[10,'Gender']='M'
data.loc[11,'Gender']='F'
data
```

```python
# Removing the duplicates
data.drop_duplicates(inplace=True)
data
```

```python
# Creating Age category new column
conditions = [
    (data['Age'] <= 25),
    (data['Age'] < 61) & (data['Age'] > 25),
    (data['Age'] >= 61) 
    ]
values=['Teen Age','Middle Age','Old Age']
data['Age_Category'] = np.select(conditions, values)
data
```

```python
# Professors logic
i=0
Age_Category=[]
for index, row in data.iterrows():
    if(row['Age'] <= 25):
    
        Age_Category.append("Teen Age")
    elif (row['Age'] < 61) & (row['Age'] > 25):
     
        Age_Category.append("Middle Age")
    else: 
       
        Age_Category.append("Old Age")
    i=i+1
data['Age_Category3']=Age_Category
data
```
